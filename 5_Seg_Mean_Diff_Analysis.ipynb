### **Estimating ideal sample size for tests**
- Evaluate the sample size required to get a significant result out of your **AB or ANOVA test**.
    - **Example**: you are working for a food delivery app to determine if a new feature which provides custom suggestions to each user based on their preferences will increase the `conversion rate` of the app. This rate measures the rate of users who "converted" or placed an order using the app. **The current CVR of the app is 12% and the stakeholders would like the new feature to increase it up to a 14%**. Given this expectation you can compute the required sample size to perform this analysis.
    
    
- For comparison of proportions of success on binary (categorical) results, the compared column must have been labelled with 0 for failure and 1 for success.

```python
TARGET_MEAN_OR_PROPORTION = 0.14
# TARGET_MEAN_OR_PROPORTION is a numeric (real) value correspondent to the mean or proportion that
# the user wants to observe. If it is a proportion, p must be input as a fraction.
# For example, if the user wants to perform a test to verify a modification of conversion rate from
# 12 to 14%, then the target proportion is 0.14, i.e., TARGET_MEAN_OR_PROPORTION = 0.14.
# If the target is a mean, any real number may be input. For instance, if user wants to verify a
# decrease of 200 to 150ºC of temperature, TARGET_MEAN_OR_PROPORTION = 150.

CURRENT_MEAN_OR_PROPORTION = None
# CURRENT_MEAN_OR_PROPORTION is a numeric value correspondent to the mean or proportion that the user
# observes now. User may directly input it or the value can be retrieved from the response column on
# the dataset. To calculate from the response on dataset, keep CURRENT_MEAN_OR_PROPORTION = None

# IN CASE OF INPUT OF PARAMETER CURRENT_MEAN_OR_PROPORTION
# If it is a proportion, p must be input as a fraction. For example, if the user wants to perform a
# test to verify a modification of conversion rate from 12 to 14%, then the
# current proportion is 0.12, i.e., CURRENT_MEAN_OR_PROPORTION = 0.12.
# If the target is a mean, any real number may be input. For instance, if user wants to verify a
# decrease of 200 to 150ºC of temperature, CURRENT_MEAN_OR_PROPORTION = 200.

CURRENT_STANDARD_DEVIATION = None
# CURRENT_STANDARD_DEVIATION = None - numeric (real) value representing
# the current standard deviation of the response variable.
# If user wants to retrieve the information from the response column on the dataset,
# CURRENT_STANDARD_DEVIATION is kept as None.
# If user wants to input the current mean value as current_mean_or_proportion (manual input)
# to check difference of mean, a real number representing the current standard deviation must be
# also input. Example: CURRENT_STANDARD_DEVIATION = 2.51
# Notice that this parameter only matters for the tests of mean difference, and is ignored
# when evaluating different proportions.
# If no value is provided, but a column and a dataset are pointed, then the function will try to
# retrieve the value from the dataframe, even if a manual value of mean is provided.


WHAT_TO_TEST = 'proportion'
# WHAT_TO_TEST = 'mean' for verifying differences between mean values.
# WHAT_TO_TEST = 'proportion' for verifying differences between proportions.

CONFIDENCE_LEVEL_PERCENT = 95
# CONFIDENCE_LEVEL_PERCENT = 95 = 95% confidence
# It is the percent of confidence for the analysis.
# Set CONFIDENCE_LEVEL_PERCENT = 90 to get 0.90 = 90% confidence in the analysis.
# Notice that, when less trust is needed, we can reduce confidence_level_pct
# to get less restrictive results.
TEST_POWER_PERCENT = 20
# TEST_POWER_PERCENT = 20 = 20% power
# It is the percent of power of the test.
# Set TEST_POWER_PERCENT = 20 to get 0.20 = 20% power in the analysis.
# Notice that, when less trust is needed, we can reduce the power to get less restrictive results.
PERFORM_TWO_SIDED_TEST = True
# PERFORM_TWO_SIDED_TEST = True. Wether or not to perform a two-sided test. Keep it
# True to use the 2-sided test, which will evaluate if there is a significant difference (higher
# or lower). This is the recommended configuration.
# If PERFORM_TWO_SIDED_TEST = False, a one-sided test will be used instead (not recommended configuration).

DATASET = None
VARIABLE_TO_ANALYZE = None
# IN CASE A MANUAL INPUT OF CURRENT MEAN OR PROPORTION WAS NOT PROVIDED:
# A dataframe object must be input as DATASET. Example: DATASET = df
# Also the column containing the current values of the variable to analyze must be provided as
# VARIABLE_TO_ANALYZE.
# DATASET is an object, so do not declare it in quotes.
# VARIABLE_TO_ANALYZE is a string, so declare it in quotes.
# Example: VARIABLE_TO_ANALYZE = 'response'
# If manual inputs are provided, you may keep DATASET = None and VARIABLE_TO_ANALYZE = None


# Variable storing the sample size was returned as required_sample_size
# Simply modify this variable on the left of equality:
required_sample_size = estimate_sample_size (target_mean_or_proportion = TARGET_MEAN_OR_PROPORTION, current_mean_or_proportion = CURRENT_MEAN_OR_PROPORTION, current_standard_deviation = CURRENT_STANDARD_DEVIATION, what_to_test = WHAT_TO_TEST, confidence_level_pct = CONFIDENCE_LEVEL_PERCENT, test_power_pct = TEST_POWER_PERCENT, perform_two_sided_test = PERFORM_TWO_SIDED_TEST, df = DATASET, variable_to_analyze = VARIABLE_TO_ANALYZE)
```
